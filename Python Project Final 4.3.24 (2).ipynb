```python
df=pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
```

# Drop unwanted columns

```python
drop_cols=['Number Scoring CR','Percent Scoring CR']
```

```python
df.drop(drop_cols,inplace=True,axis=1)
```

```python
df.head()
```

# Add a new column for Borough

This is based on School DBN (Note: DBN 08X530 as example, X indicates a borough number. Borough number is always at the 3rd position in the DBN format)(K=Brooklyn, M=Manhattan,Q=Queens,R=Staten Island, X=Bronx)

```python
df['Borough']=df['School DBN'].str[2]
```

```python
df
```

# Replace invalid values 's','na','--'

```python
missing_values=['s','na','--']
```

```python
df.replace(missing_values,float('NaN'),inplace=True)
```

```python
df
```

# Check missing values under 'Regents Exam' column

```python
missing_exam_rows=df[df['Regents Exam'].isnull()]
```

```python
missing_exam_rows
```

# Remove invalid values 'NaN'

```python
df_cleaned=df.dropna()
```

```python
df_cleaned
```

```python
df.dtypes
```

# Convert Data Type from 'Object' to 'Numeric'

```python
columns_to_covert=['Mean Score','Number Scoring Below 65','Percent Scoring Below 65','Number Scoring 65 or Above','Percent Scoring 65 or Above','Number Scoring 80 or Above','Percent Scoring 80 or Above']
```

```python
df_cleaned[columns_to_covert]=df_cleaned[columns_to_covert].apply(pd.to_numeric,errors='coerce')
```

```python
df.dtypes
```

# Analysis by School Level

```python
mean_scores_by_level=df_cleaned.groupby('School Level')['Mean Score'].mean().sort_values(ascending=False)
```

# Using the above Lehman College as the Comparison Point

Filter the dataset for the year 2017

```python
df_2017=df_cleaned[df_cleaned['Year']==2017]
```

```python
df_2017
```

```python
df_high_schools=df_2017[df_2017['School Level']=='High school']
```

```python
df_high_schools
```

Filter Dataset by Regents Exam-U.S. History and Government

```python
us_history_regent_exam_avg=df_high_schools[df_high_schools['Regents Exam']=='U.S. History and Government']
```

```python
us_history_regent_exam_avg
```

Filter down to High School of American Studies at Lehman College

```python
Lehman_us_history_regent_exam_score=us_history_regent_exam_avg[us_history_regent_exam_avg['School Name']=='High School of American Studies at Lehman College']
```

```python
Lehman_us_history_regent_exam_score
```

# Lehman High School U.S. History Exam Mean Score

```python
Lehman_us_history_exam_mean_score=Lehman_us_history_regent_exam_score['Mean Score'].mean()
```

```python
Lehman_us_history_exam_mean_score
```

# Filter down to Queens high schools & exam avg

Filter down to Queens high schools only using Borough column

```python
df_queens_high_schools=us_history_regent_exam_avg[us_history_regent_exam_avg['Borough']=='Q']
```

```python
df_queens_high_schools
```

Queens high schools' Mean Scores on 'U.S. History and Government'

```python
df_queens_high_schools_mean_score=df_queens_high_schools.groupby(['Regents Exam'])['Mean Score'].mean()
```

```python
df_queens_high_schools_mean_score
```

# Plot the comparison 

Comparison is based on Regents Exam-"U.S. History & Government' between mean scores of High School of American Studies at Lehman College & all Queens high schools

# Plotting

```python
import matplotlib.pyplot as plt
```

Mean Score for U.S. History & Government at High School of American Studies at Lehman College

```python
Lehman_us_history_exam_mean_score=98.35
```

Mean Score for U.S. History & Government at all Queens High Schools

```python
df_queens_high_schools_mean_score=73.57
```

Plotting the Mean Scores

```python
plt.bar('Queens High Schools',df_queens_high_schools_mean_score, color='yellow')
plt.bar('High School of American Studies at Lehman College',Lehman_us_history_exam_mean_score,color='green')
plt.xlabel('Schools',fontsize=20)
plt.ylabel('Mean Score',fontsize=20)
plt.title('Mean Scores for Regents Exam-U.S. History & Government',fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

# Conclusion Statement

Based on the dataset, we can conclude that the High School of American Studies at Lehman College tends to have a higher mean score on the Regents Exam in U.S. History & Government compared to the average mean scores of all Queens high schools. This shows that students at the Lehman college, on average, perform better in this particular subject compared to students at Queens high schools.
